# Egg Price Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set_style("darkgrid")
```

# Data cleaning and Manipulation

```python
df=pd.read_excel("Egg Rate Compare Yearly - V1.xlsx")
```

```python
df.head()
```

```python
df.columns=['Date',"Rate"]
```

```python
df.head()
```

```python
df=df.iloc[1:,:]
```

```python
df.tail(20)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.replace("-",0)
df=df.fillna(0)
```

```python
df['Rate']=df["Rate"].astype(int)
df['Date']=pd.to_datetime(df["Date"],"DD-MM--YY")
```

```python
df.info()
```

```python
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['Date']).year
df['month'] = pd.DatetimeIndex(df['Date']).month
df['week'] = pd.DatetimeIndex(df['Date']).day_name()
df['day'] = df['Date'].dt.day
```

```python
df.head(10)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(df["Rate"],palette='PuBu')
plt.show()
```

```python
q1=df['Rate'].quantile(0.25)
q3=df["Rate"].quantile(0.75)
IQR=q3-q1
ll=q1-1.5*IQR
ul=q3+1.5*IQR
```

```python
df=df[(df["Rate"]>ll) & (df["Rate"]<ul)]
```

```python
df[(df["Rate"]>450) & (df["Rate"]<550)]
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(df["Rate"],palette='PuBu')
plt.show()
```

# Model Building

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr())
```

```python
df_model=pd.get_dummies(df,drop_first=True)
```

```python
df_model.head()
```

```python
X=df_model.drop(['Date','Rate'],axis=1)
y=df_model['Rate']
```

```python
#spliiting train_test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score,mean_squared_error
```

```python
pipeline_ada=Pipeline([('scalar1',StandardScaler()),
                     
                     ('lr_classifier',AdaBoostRegressor(random_state=0))])
```

```python
pipeline_dt=Pipeline([('scalar2',StandardScaler()),
                     
                     ('dt_classifier',DecisionTreeRegressor())])
```

```python
pipeline_randomforest=Pipeline([('scalar3',StandardScaler()),
                     
                     ('rf_classifier',RandomForestRegressor(random_state=0))])
```

```python
## LEts make the list of pipelines
pipelines = [pipeline_ada, pipeline_dt, pipeline_randomforest]
```

```python
best_accuracy=0.0
best_classifier=0
best_pipeline=""
```

```python
# Dictionary of pipelines and classifier types for ease of reference
pipe_dict = {0: 'ADA', 1: 'Decision Tree', 2: 'RandomForest'}

# Fit the pipelines
for pipe in pipelines:
    pipe.fit(X_train, y_train)
```

```python
for i,model in enumerate(pipelines):
    print("{} Test Accuracy: {}".format(pipe_dict[i],model.score(X_test,y_test)))
```

```python
# Create a pipeline
pipe = Pipeline([ ('scalar1',StandardScaler()),
                  ("classifier", RandomForestRegressor())
                 
                ])
                
                
# Create dictionary with candidate learning algorithms and their hyperparameters
grid_param = [ 
                 {"classifier": [RandomForestRegressor()],
                  
                 "classifier__n_estimators": [215,216],
                 "classifier__max_depth":[35,37,39],
                 #"classifier__min_samples_leaf":[2,4,5],
                 "classifier__max_leaf_nodes": [135,138,140,150],
                  "classifier__max_features":  ["auto", "sqrt", "log2"]
                 }
            
              ]
# create a gridsearch of the pipeline, the fit the best model
gridsearch = GridSearchCV(pipe, grid_param, cv=KFold(n_splits=5),verbose=0,n_jobs=-1) # Fit grid search
best_model = gridsearch.fit(X_train,y_train)
```

```python
print(best_model.best_estimator_)
print("The mean accuracy of the model is:",best_model.score(X_test,y_test))
```

```python
pipeline_randomforest.fit(X_train, y_train)
y_pred= pipeline_randomforest.predict(X_test)
```

```python
r2_score(y_pred,y_test)
```

```python
mean_squared_error(y_pred,y_test)
```
